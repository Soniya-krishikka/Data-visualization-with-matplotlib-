```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\subal\\Downloads\\householdtask3.csv')
```

```python
data.head(10)
```

```python
#Scatter plot with year against own 

plt.scatter(data ['year'], data ['own'])

#Adding title to the plot 

plt.title("Scatter Plot")

#Setting the x and y Lobels 

plt.xlabel('year') 
plt.ylabel('own')
           
#Showing the chart
           
plt.show()
```

```python
#line chart with age against size 

plt.plot(data ['age'])
plt.plot(data ['size'])

#Adding title to the plot 

plt.title("line chart")

#Setting the x and y Lobels 

plt.xlabel('age') 
plt.ylabel('size')
           
#Showing the chart
           
plt.show()
```

```python
#bar chart with year against expenditure

plt.bar(data ['year'], data ['expenditure'])

#Adding title to the chart 

plt.title("bar chart")

#Setting the x and y Lobels 

plt.xlabel('year') 
plt.ylabel('expenditure')
           
#Showing the chart
           
plt.show()
```

```python
#Histogram

plt.hist(data['income'])

#Adding title to the chart 

plt.title("Histogram")

#Showing the chart

plt.show()
```
